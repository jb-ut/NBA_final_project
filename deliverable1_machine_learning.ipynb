```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
```

```python
#  Import and read the nba data.
file_name = 'Resources/game_edited_test.csv'
df = pd.read_csv(file_name)
df.head()
```

```python
le = LabelEncoder()
df['TEAM_ID_HOME'] = le.fit_transform(df['TEAM_ID_HOME'])
df['TEAM_ID_AWAY'] = le.fit_transform(df['TEAM_ID_AWAY'])
df.head()
```

```python
#drop non-numerical data
df = df.dropna()
drop_df = df.drop(columns=['HOME_W'], axis=1)
drop_df.head()
```

```python
# split into features and target arrays
y = df['HOME_W'].values
X = drop_df.values
```

```python
# create standard scalar instance
scaler = MinMaxScaler()
```

```python
#split preprocesed data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 10)
```

```python
#fit the standardscaler
X_scaler = scaler.fit(X_train)
```

```python
#scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# define the model
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 8
hidden_nodes_layer2 = 5
#hidden_nodes_layer3 = 2

nn = tf.keras.models.Sequential()

#first hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
#nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="sigmoid"))

#output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

#check structure
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
